a. Consider the mathematical statement $\forall x,y\in \mathbb{R} : (x\neq y) \implies (\exists z\in\mathbb{R} : (x < z < y) \vee (y < z < x))$.

a.1 Translate the mathematical statement into a complete English sentence, including proper capitalization and punctuation; **include no mathematical symbols in your English translation**.  

a.2 Is this statement true or false?

a.3 Can you prove (or provide a counterexample to substantiate) your answer to (a.2)?


b. Consider the polynomial expression $a x^2 + b x + c$.

c. Consider the complex number $z = r e^{\theta j}\in\mathbb{C}$, where $r,\theta\in\mathbb{R}$.

d. Consider the functions $f : X \rightarrow Y$, $g : Y \rightarrow Z$.
