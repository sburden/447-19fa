# problem (from Fa18 midterm exam)

Consider the scalar nonlinear differential equation

$$ (NL)\quad \dot{x} = f(x,u) = - r x \left(1 - \frac{x}{c} + u\right) $$

that models dynamics of a population $x > 0$ with rate parameter $r > 0$, carrying capacity $c > 0$, and input $u$.  Note that if the input is constant, $u(t) = u_e > 0$, the population levels $0$ (zero) and $x_e = c(1 + u_e)$ are equilibria for $(NL)$.


## subproblem a.

Graph $f(x,u_e)$ on an interval containing both equilibria, and indicate whether each equilibrium is stable.


## subproblem b.

Now $(NL)$ is linearized around the positive equilibrium $x_e$ to obtain

$$ (L)\quad \delta\dot{x} = a\,\delta x + b\,\delta u $$

where $a,b > 0$ are such that $x \approx x_e + \delta x$ when $u = u_e + \delta u$.

(**Note:**  you **don't** need to determine the value of the variables $a,b$ to solve this problem.)

Suppose proportional feedback $\delta u = -k\,\delta x$ is applied to $(L)$.  Use the Routh-Hurwitz stability criterion to determine the greatest lower bound on $k > 0$ that ensures $(L)$ is asymptotically stable in terms of $a,b$.

## subproblem e.
